# A few examples of plotting with loops

We'd like to plot the function $f(k, x)$ vs. $x$ for several values of $k$.

```python
def f(k, x):
    '''Some 2-parameter function'''
    return np.sinc(k * x)
```

```python
# We'll use the same k and x values for all the plots:
k_array = np.arange(0.1, 0.5, 0.1)
x = np.linspace(-10, 10)
```

## Method 1: Generate the data on the fly inside the loop:

```python
# Method 1: Generate the data on the fly inside the loop:

fig1, ax1 = plt.subplots()
for k_i in k_array:
    ax1.plot( x, f(k_i,x), label=r'$k={:.1f}$'.format(k_i) )

# Label the plot, etc.
ax1.legend()
ax1.set_xlim(x[0], x[-1])
ax1.set_ylabel(r'$f(k,x)$')
ax1.set_xlabel(r'$x$')
ax1.set_title('Figure 1');
```

# Method 2: Pre-compute the data and store in a list
* You might do this if generating the data takes some time; you don't want to wait each time you tweak your plots. 

```python
# Method 2: Pre-compute the data and store in an array

# Initialize the data_list
f_list = []

# Commpute and store data in rows
for k_i in k_array:
    f_list.append( f(k_i, x) )


# Below might be in a new cell
##############################

#Unpack the data and plot
fig2, ax2 = plt.subplots()
for k_i, f_i in zip(k_array, f_list):
    ax2.plot( x, f_i, label=r'$k={:.1f}$'.format(k_i) )

# Label the plot, etc.
ax2.legend()
ax2.set_xlim(x[0], x[-1])
ax2.set_ylabel(r'$f(k,x)$')
ax2.set_xlabel(r'$x$')
ax2.set_title('Figure 2');
```

# Method 3: Pre-compute the data and store in a numpy array
* You might do this if if you need the data later on for other analysis and want the data formatted as an array.

```python
# Method 3: Pre-compute the data and store in an array

# Initialize the data_array
f_array = np.zeros( (k_array.size, x.size) )

# Commpute and store data in rows
for i, k_i in enumerate(k_array):
    f_array[i,:] = f(k_i, x)


# Below might be in a new cell
##############################

# Unpack the data and plot
fig2, ax2 = plt.subplots()
for k_i, f_i in zip(k_array, f_array):
    ax2.plot( x, f_i, label=r'$k={:.1f}$'.format(k_i) )

# Label the plot, etc.
ax2.legend()
ax2.set_xlim(x[0], x[-1])
ax2.set_ylabel(r'$f(k,x)$')
ax2.set_xlabel(r'$x$')
ax2.set_title('Figure 3');
```

# Method 4: Pre-compute the data and store in a dictionary
* [Python dictionaries](https://docs.python.org/3/tutorial/datastructures.html#dictionaries) are very handy!

```python
# Method 3: Pre-compute the data and store in a dictionary

# Initialize the dictionary
f_dict = {}

# Commpute and store data
for k_i in k_array:
    f_dict[k_i] = f(k_i, x)


# Below might be in a new cell
##############################

# Unpack the data and plot
fig2, ax2 = plt.subplots()
for k_i, f_i in f_dict.items():
    ax2.plot( x, f_i, label=r'$k={:.1f}$'.format(k_i) )

# Label the plot, etc.
ax2.legend()
ax2.set_xlim(x[0], x[-1])
ax2.set_ylabel(r'$f(k,x)$')
ax2.set_xlabel(r'$x$')
ax2.set_title('Figure 3');
```
