```python
#delta x = 0.1

import numpy as np
import matplotlib.pyplot as plt

data_01 = np.loadtxt("myPsi_01.dat")
x_01 = data_01[:, 0]
psi_01 = data_01[:, 1]
wav_01 = np.cos((np.pi * x_01)/2)

plt.plot(x_01, np.abs(psi_01-wav_01))
plt.yscale('log')
```
